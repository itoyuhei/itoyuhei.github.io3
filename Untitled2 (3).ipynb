```python
import ipywidgets as widgets
from IPython.display import display, Image
from IPython.core.display import HTML

todo_list = {}

def create_todo_widget():
    date_input = widgets.DatePicker(description='日付')
    todo_input = widgets.Text(description='予定')
    add_button = widgets.Button(description='追加')
    output = widgets.Output(layout={'width': '50%', 'margin': '0 auto'})

    def add_todo_item(button):
        date = date_input.value.strftime('%Y-%m-%d')
        todo = todo_input.value.strip()

        if date and todo:
            if date in todo_list:
                index = len(todo_list[date]) + 1
            else:
                index = 1

            todo_item = {'日付': date, '予定': todo}
            todo_list.setdefault(date, {})[index] = todo_item
            with output:
                print(f'追加された予定: {todo}')
            update_select_options()

            if "うんこ" in todo_item["予定"]:
                image_url = "https://hiyokoyarou.com/wp-content/uploads/2017/04/unchikun-1024x1024.png"
                display(HTML(f'<div style="text-align:center;"><img src="{image_url}" width="400px"></div>'))

        date_input.value = None
        todo_input.value = ''

    add_button.on_click(add_todo_item)

    return widgets.VBox([date_input, todo_input, add_button, output])

def update_select_options():
    todo_select.options = [(f'{date} - {index}', date, index)
                           for date, items in sorted(todo_list.items())
                           for index in sorted(items.keys())]

def select_todo_item(change):
    selected_item = change.new
    if selected_item:
        todo_output.clear_output()  # 選択結果を消去
        date, index = selected_item[1], selected_item[2]
        todo_item = todo_list[date][index]
        with todo_output:
            display(HTML(f'<h3>日付: {todo_item["日付"]}</h3>'))
            display(HTML(f'<h3>予定: {todo_item["予定"]}</h3>'))

    else:
        with todo_output:
            display(HTML('<strong>予定を選択してください</strong>'))

            # 追加された予定の表示結果に対して画像があれば表示
            for date, items in sorted(todo_list.items()):
                for index, todo_item in sorted(items.items()):
                    if "うんこ" in todo_item["予定"]:
                        image_url = "https://hiyokoyarou.com/wp-content/uploads/2017/04/unchikun-1024x1024.png"
                        display(HTML(f'<div style="text-align:center;"><img src="{image_url}" width="400px"></div>'))

new_button = widgets.Button(description='新規作成')
todo_select = widgets.Select(description='Todoリスト', options=[], layout={'width': '50%'})
todo_output = widgets.Output(layout={'width': '50%', 'margin': '0 auto'})

new_button.on_click(lambda _: todo_output.clear_output())
new_button.on_click(lambda _: todo_output.append_display_data(create_todo_widget()))

todo_select.observe(select_todo_item, names='value')

display(widgets.VBox([new_button, widgets.HBox([todo_select, todo_output])]))
```
